#### Adding new column duration

```python
df['starttime']=pd.to_datetime(df['starttime'])
df['endtime']=pd.to_datetime(df['endtime'])
```

### There are some end date marked as same date, lets try to fix it by making the day as next date



```python
df1=df[df['endtime']<df['starttime']] 
df2=df[df['endtime']>df['starttime']] # store temporarily
df1['endtime']=pd.DatetimeIndex(df1['endtime']) + pd.DateOffset(1)
```

```python
df=pd.concat([df1, df2])
```

```python
df['duration']=df['endtime'].sub(df['starttime'],axis=0)
```

```python
# Lets try to see if the problem is fixed; this next line should not return any rows
df[df['endtime']<df['starttime']]
```

```python
#df2['newendtime']=pd.DatetimeIndex(df[df['endtime']>df['starttime']]['endtime']) + pd.DateOffset(1)
#df2=df2.drop('endtime',axis=1).rename(columns={'newendtime':'endtime'})
```

```python
df['sortedbyduration'] = df['duration'].rank()
df=df.drop("sortedbyduration",axis=1)
```

```python
df=df.sort_values(by=['duration'],ascending=False)
print "printing the dataframe after sorting by duration : \n"
df
```

```python
df['duration_sec'] = df['duration'].apply(lambda x: x / np.timedelta64(1, 's'))
```
